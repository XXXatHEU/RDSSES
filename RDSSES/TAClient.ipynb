```python
import time
import sys
import pickle
import socket
from charm.toolbox.pairinggroup import PairingGroup, G1,GT,G2, ZR, pair as e,extract_key
from hashlib import sha256, md5
from util1 import obj2bytes, read_word_as_bytes, bxor
import import_ipynb
from Hash_Util import *
class TAClient:
    def __init__(self, addr):
        self.addr = addr
    def Setup(self, U,Mapping):
        conn = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        conn.connect(self.addr)
        conn.send(pickle.dumps((0 , (U,Mapping))))
        data1 = b""
        while True:
            packet = conn.recv(4096)
            if packet is not None:
                data1 += packet
            if packet is None or len(packet) < 4096:
                break
        data = pickle.loads(data1)
        conn.close()
        return data
                  
    def UpdateKeyGen(self, attr):
        conn = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        conn.connect(self.addr)
        conn.send(pickle.dumps((2,attr)))
        data1 = b""
        while True:
            packet = conn.recv(4096)
            if packet is not None:
                data1 += packet
            if packet is None or len(packet) < 4096:
                break
        data = pickle.loads(data1)
        conn.close()
        return data
    def UpdateCSP_CTList(self, PK,RL ):
        conn = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        conn.connect(self.addr)
        conn.send(pickle.dumps((6,(PK,RL))))
        data1 = b""
        while True:
            packet = conn.recv(4096)
            if packet is not None:
                data1 += packet
            if packet is None or len(packet) < 4096:
                break
        data = pickle.loads(data1)
        conn.close()
        return data
    

MAXINT = sys.maxsize
HOST = 'localhost'
PORT = 50098
CSP_PORT = 50068
group = PairingGroup('MNT159')

if __name__ == "__main__":
    #HOST = sys.argv[1]
    #PORT = int(sys.argv[2])
    # HOST = 'localhost'
    # PORT = 50060
    PK = None
    MSK = None
    client_obj = TAClient((HOST, PORT))
    client2_obj = TAClient((HOST, CSP_PORT))
    print('欢迎进入TA操作界面用odxt可搜索加密系统')
    while True:
        option = input('请输入选项前面的数字：1. 初始化TA ; 2.属性撤销  ; exit退出')
        if option == '1':
            
            U = []
            Mapping = {}
            with open('30.txt', 'r') as f:
                U = f.readlines()
                U = [line.strip() for line in U]
            result_dict = {}
            with open('30_map.txt', 'r') as f:
                for line in f:
                    line = line.strip()  # 去除行末的换行符
                    if ':' in line:
                        key, value = line.split(':')
                        Mapping[value.strip()] = key.strip()
            start_time = time.time()  # 记录程序开始时间
            PK_MSK = client_obj.Setup(U,Mapping)
            
            PK = PK_MSK['PK']
            PK[3] = group.deserialize(PK[3])
            PK[7] = group.deserialize(PK[7])
            PK[8] = group.deserialize(PK[8])
            for key in PK[9].keys():
                PK[9][key] = group.deserialize(PK[9][key])
            PK[10] = group.deserialize(PK[10])    
            MSK = PK_MSK['MSK']
            MSK[0] = group.deserialize(MSK[0])
            MSK[1] = group.deserialize(MSK[1])
            for key in MSK[2].keys():
                MSK[2][key] = group.deserialize(MSK[2][key])
            #print(PK)
            #print(MSK)
            end_time = time.time()  # 记录程序结束时间
            total_time = (end_time - start_time) * 1000  # 计算程序运行时间（毫秒）
            print(f"程序运行时间为：{total_time:.2f}毫秒")
        elif option == '2':
            attr = input('请输入要撤销的属性>')
            data = client_obj.UpdateKeyGen(attr)
            PK_s = data['PK']
            MSK_s = data['MSK']
            RL_s = data['RL']
            PK[3] = group.deserialize(PK_s[3])
            PK[7] = group.deserialize(PK_s[7])
            PK[8] = group.deserialize(PK_s[8])
            for key in PK_s[9].keys():
                PK[9][key] = group.deserialize(PK_s[9][key])
            PK[10] = group.deserialize(PK_s[10])    
            MSK[0] = group.deserialize(MSK_s[0])
            MSK[1] = group.deserialize(MSK_s[1])
            for key in MSK_s[2].keys():
                MSK[2][key] = group.deserialize(MSK_s[2][key])
            print(PK)
            print(MSK)
            #将PK , MSK 广播出去
            #将RL,PK传输到CSP
            client2_obj.UpdateCSP_CTList(PK_s,RL_s)
            print("CSP撤销属性成功")
            
            print(data)
        elif option == 'exit':
            print('退出！')
            break
        else:
            print('输入无效，请重新输入')
```
